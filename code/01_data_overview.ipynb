# 가설 설정

### 가설 1 : D1 retention

- 귀무가설 (H0) : gate 40의 D1 retention은 gate 30의 D1 retention보다 작거나 같다.
- 대립가설 (H1) : gate 40의 D1 retention이 gate 30의 D1 retention보다 크다.

### 가설 2 : D7 retention

- 귀무가설 (H0) : gate 40의 D7 retention은 gate 30의 D7 retention보다 작거나 같다.
- 대립가설 (H1) : gate 40의 D7 retention이 gate 30의 D7 retention보다 크다.

---
# KPI 정의

- Primary KPI
    > D7 retention
    설치 후 7일 이내 최소 1회 이상 재접속한 유저의 비율

- Secondary KPI
    > D1 retention:
    설치 후 1일 이내 재접속한 유저의 비율

- Third KPI
    > 하루 평균 플레이 스테이지 수:
    설치 후 7일 동안 유저당 평균 플레이 스테이지 수
